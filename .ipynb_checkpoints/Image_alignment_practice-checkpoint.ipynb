## Function that can use two different methods to make an image grayscale
Note: Image.conver("L") from PIL will do the same thing as rgb2gray(a, method = "wgt")

```python
def rgb2gray (a, method = "avg"):                   
    s = np.shape(a)
    bw = np.zeros((s[0],s[1]))
    if method == "avg":
        w = np.array((1,1,1))/3.
        print ('using avg')
    elif method == "wgt":
        w = np.array((0.299,0.587,0.114))
        print('using wgt')
    else: 
        w = np.array((1,1,1))/3.  
        print ("defaulting to wgt")
    bw = a[:,:,0]*w[0]+a[:,:,1]*w[1]+a[:,:,2]*w[2]
    return bw

bw = rgb2gray(img1,"wgt")
plt.imshow(bw, cmap = 'gray')
plt.colorbar()
```

## Function that will calculate the Mean Squared Error (MSE) and Structural Similarity Index (SSIM)
SSIM is more precise at the scales usable.  
For MSE,  perfect similarity = 0 MSE, values >=1 show decreasing similarity, the higher the value the less similar. 
For SSIM, the scale is -1 to 1 with one being perfect similarity.

```python
# Import modules
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import mean_squared_error as mse 
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# define the function as taking two images for an argument
def error_metric(ima, imb):
    # Open images
    ima = np.asarray(Image.open(ima))
    imb = np.asarray(Image.open(imb))
    # use scikit-image to get the mean squared error and structural similarity index
    m = mse(ima,imb)
    print ("Mean Squared Error = {}".format(m))
    s = ssim(ima, imb, multichannel=True)
    print ("Structural Similarity Index = {} ".format(s))
    return
```

### Function that returns arrays of the norm of the elementwise difference of two arrays 

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def imgarray_dif(img1, img2):
    # read images
    img1 = plt.imread(img1)
    img2 = plt.imread(img2)
    # convert images into 2D grayscale arrays
    img1 = rgb2gray(img1, "wgt") 
    img2 = rgb2gray(img2, "wgt")
    # compare the arrays
    # Manhatten Norm
    m_norm = sum( abs(img1-img2) ) / img1.size
    # Normalized Cross-Correlation [wikipedia page on formulas](https://en.wikipedia.org/wiki/Cross-correlation#Normalized_cross-correlation)
    cc_norm = sum( (img1 - np.mean(img1)) * (img2 - np.mean(img2)) ) / ((img1.size - 1) * np.std(img1) * np.std(img2) )
    # show values
    print("Manhatten Norm Array: {}".format(m_norm))
    print("Normalized Cross Correlation Array: {}".format(cc_norm))
    return

    
    
    
    
```

# Three Stooges Method
Method: Manually entering values to translate an image by x,y pixels until the error between the image and its reference is acceptable.  

Incorporating the error_metric(), this method can be done by writing a function that applies a x,y translation to an image and then returns the Mean Squared Error and Structural Similarity Index of the translated image. This will help speed up the process of manually entering translation coordinates until an acceptable error metric is achieved.   

Sources: [This](https://stackoverflow.com/questions/54274185/shifting-an-image-by-x-pixels-to-left-while-maintaining-the-original-shape/54274222) stackoverflow page where the first answer provides some info about image translation. 

```python
# import libraries 
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import mean_squared_error as mse 
import numpy as np
from PIL import Image
import cv2


# define function
def stooge(im1, im2, x_change, y_change):
    
    # Find the intital difference of error between the two images. 
    print ('Initial error metrics:')
     # Open images
    ima = cv2.imread(im1,0)
    imb = cv2.imread(im2,0)
    # use scikit-image to get the mean squared error and structural similarity index
    m = mse(ima,imb)
    print ("Mean Squared Error = {}".format(m))
    s = ssim(ima, imb)
    print ("Structural Similarity Index = {} ".format(s))
    
    # shfit the image by creating a transformation matrix
    trans_matrix = np.float32([[1,0,x_change], [0,1,y_change]])
    rows, cols = imb.shape
    shifted_image = cv2.warpAffine(imb, trans_matrix, (cols,rows))
    
    # Show new error metrics
    print ('Image translation error metrics:')
    m = mse(ima, shifted_image)
    print ("Mean Squared Error = {}".format(m))
    s = ssim(ima, shifted_image)
    print ("Structural Similarity Index = {} ".format(s))
    
    horiz = np.concatenate((ima, shifted_image), axis = 1)
    vert = np.concatenate((ima, shifted_image), axis = 0)
    
    cv2.imshow('HORIZONTAL', horiz)
    cv2.imshow('VERTICAL', vert)
  
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    return
```

```python
stooge("quiz_images/small1.jpg", "quiz_images/small1_shifted.jpg", 40, 10)
```

### My attempt to apply the method to GLERL images

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy import signal, ndimage, misc

def plot_corr(i1,i2):
    # load images and convert to grayscale
    img1 = Image.open(i1).convert('L')
    img2 = Image.open(i2).convert('L')
    
    # turn the PIL images into numpy arrays and resize to avoid a memory error
    a1 = np.array(img1.resize((img1.width// 2, img1.height // 2)))
    a2 = np.array(img2.resize((img2.width// 2, img2.height // 2)))
    
    # Run correlation commands
    if len(np.shape(a1)) == 2 and len(np.shape(a2)) == 2:
        corr = signal.correlate2d(a1, a2, mode = 'same') #this command will run forever.
    else:
        print ('Must used two 2D arrays for correlation')
   
    # compute position of maximum correlation
    y, x = np.unravel_index(np.argmax(corr), corr.shape)
    print (x,y)
    
    # plot results
    y2, x2 = np.array(a2.shape) // 2
    print (x2, y2)

    fig, (ax_a1, ax_a2, ax_corr) = plt.subplots(1, 3, figsize=(15, 5))
    im = ax_a1.imshow(a1, cmap='gray')
    ax_a1.set_title('img1')
    ax_a2.imshow(a2, cmap='gray')
    ax_a2.set_title('img2')
    im = ax_corr.imshow(corr, cmap='viridis')
    ax_corr.set_title('Cross-correlation')
    ax_a1.plot(x, y, 'ro')
    ax_a2.plot(x2, y2, 'go')
    ax_corr.plot(x, y, 'ro')
    fig.show()
```
