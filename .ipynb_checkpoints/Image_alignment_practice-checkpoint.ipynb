## Function that can use two different methods to make an image grayscale
Note: Image.conver("L") from PIL will do the same thing as rgb2gray(a, method = "wgt")

```python
def rgb2gray (a, method = "avg"):                   
    s = np.shape(a)
    bw = np.zeros((s[0],s[1]))
    if method == "avg":
        w = np.array((1,1,1))/3.
        print ('using avg')
    elif method == "wgt":
        w = np.array((0.299,0.587,0.114))
        print('using wgt')
    else: 
        w = np.array((1,1,1))/3.  
        print ("defaulting to wgt")
    bw = a[:,:,0]*w[0]+a[:,:,1]*w[1]+a[:,:,2]*w[2]
    return bw

bw = rgb2gray(img1,"wgt")
plt.imshow(gray2-bw, cmap = 'gray')
plt.colorbar()
```

## Function that will calculate the Mean Squared Error (MSE) and Structural Similarity Index (SSIM)
SSIM is more precise at the scales usable.  
For MSE,  perfect similarity = 0 MSE, values >=1 show decreasing similarity, the higher the value the less similar. 
For SSIM, the scale is -1 to 1 with one being perfect similarity.

```python
# Import modules
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import mean_squared_error as mse 
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# define the function as taking two images for an argument
def error_metric(ima, imb):
    # convert images to grayscale numpy arrays
    ima = np.asarray(Image.open(ima).convert("L"))
    imb = np.asarray(Image.open(imb).convert("L"))
    # use scikit-image to get the mean squared error and structural similarity index
    m = mse(ima,imb)
    print ("Mean Squared Error = {}".format(m))
    s = ssim(ima, imb)
    print ("Structural Similarity Index = {} ".format(s))
    return
```

### Function that returns the norm of the elementwise difference of two arrays 

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def imgarray_dif(img1, img2):
    # read images
    img1 = plt.imread(img1)
    img2 = plt.imread(img2)
    # convert images into 2D grayscale arrays
    img1 = rgb2gray(img1, "wgt") 
    img2 = rgb2gray(img2, "wgt")
    # compare the arrays
    # Manhatten Norm
    m_norm = sum( abs(img1-img2) ) / img1.size
    # Normalized Cross-Correlation [wikipedia page on formulas](https://en.wikipedia.org/wiki/Cross-correlation#Normalized_cross-correlation)
    cc_norm = sum( (img1 - np.mean(img1)) * (img2 - np.mean(img2)) ) / ((img1.size - 1) * np.std(img1) * np.std(img2) )
    # show values
    print("Manhatten Norm Array: {}".format(m_norm))
    print("Normalized Cross Correlation Array: {}".format(cc_norm))
    return

    
    
    
    
```
